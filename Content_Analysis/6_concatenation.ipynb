# **Xiao's labeled videos**

Extract MER

```python
f1=open('Content Features Table (Labeled data - Xiao)/labeled_content_mer_cnt.txt')
mer1=[x.strip('\n').split(' ') for x in f1.readlines()]
MER1={}
for x in mer1:
  MER1[x[0][:-4]]=int(x[1])

f2=open('Metadata (Labeled data - Xiao) & Features Table/labeled_description_mer_cnt.txt')
mer2=[x.strip('\n').split(' ') for x in f2.readlines()]
MER2={}
for x in mer2:
  MER2[x[0][:-4]]=int(x[1])
  
```

Join metadata and content view table

```python
labeled_content_pd_raw=pd.read_csv('Content Features Table (Labeled data - Xiao)/labeled_video_raw_content.csv')
labeled_meta_pd_raw=pd.read_csv('Metadata (Labeled data - Xiao) & Features Table/features_labeled_videos.csv')
labeled_content_pd_raw=labeled_content_pd_raw.rename(columns={"readability": "narrative_readability"})

total_ids=list(labeled_content_pd_raw['video_ID'])
labeled_m_id=list(labeled_meta_pd_raw['id'])
labeled_vid_table={}
print(total_ids)
for i,v in enumerate(total_ids):
  labeled_vid_table[v]={}
  for q in labeled_content_pd_raw.columns[1:]:
    labeled_vid_table[v][q]=labeled_content_pd_raw.loc[i][q]
  if v in labeled_m_id:
    j=labeled_m_id.index(v)
    for q in labeled_meta_pd_raw.columns[1:]:
      labeled_vid_table[v][q]=labeled_meta_pd_raw.loc[j][q]
  if v in MER1.keys():
    labeled_vid_table[v]['transcript_MER_terms']=MER1[v]

  if v in MER2.keys():
    labeled_vid_table[v]['description_MER_terms']=MER2[v]

print(labeled_vid_table)
```

```python
cols=labeled_vid_table['-71ld0iqAq8'].keys()
idx=labeled_vid_table.keys()
labeled_vid_table_pd=pd.DataFrame(columns=cols,index=idx)

for i,id in enumerate(idx):
  for col in cols:
    if col in labeled_vid_table[id].keys():
      labeled_vid_table_pd.iloc[i][col]=labeled_vid_table[id][col]

labeled_vid_table_pd=labeled_vid_table_pd.reset_index()
#
#json file 
labeled_vid_table_pd=labeled_vid_table_pd.rename(columns={'index':'video_ID'})
labeled_vid_table_pd=labeled_vid_table_pd.set_index('video_ID')
#labeled_vid_table_pd=labeled_vid_table_pd.drop(columns='test')
labeled_vid_table_pd
```

```python
c=list(labeled_vid_table_pd.columns)
c.sort()
labeled_vid_table_pd=labeled_vid_table_pd[c]

labeled_vid_table_pd.to_csv('final_csv/labeled_video_data.csv')
```
