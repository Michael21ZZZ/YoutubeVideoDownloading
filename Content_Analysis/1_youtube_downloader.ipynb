```python
# coding: utf-8
import os
import time
import glob
from urllib.error import HTTPError
```

```python
from collections import Counter
BASE_URL = 'https://www.youtube.com/watch?v='


def get_downloadURLSet(filename):
    URL_set = set()
    with open(filename) as f:
        for eachURL in f.readlines()[3:]:
            eachURL = eachURL.rstrip('\n')
            URL_set.add(BASE_URL + eachURL)
            # print(eachURL)
    return URL_set


def list_current_IDs(directory):
  raw=glob.glob(directory+"/*.mp4")
  res=[]
  for r in raw:
    res.append(r[len(directory+'/'):-4])
  return res

def download_video(downloadURLSet):
    if len(downloadURLSet) == 0:
        print("no video to download")
        return
    ids=list_current_IDs('videos_trial')
    total_sec=0
    count=0
    for downloadURL in downloadURLSet:
        if downloadURL.strip(BASE_URL) not in ids:
          print(downloadURL)
          t=time.time()
          try:
            yt = YouTube(downloadURL)
          except:
            print("Some thing wrong about the authority!")
            continue
          try:
            name = yt.title
            print("Now is loading %s------------>" % name)
            stream = yt.streams.filter(file_extension='mp4').first()
            #  
            stream.download('videos_trial')
            os.rename(os.path.join('videos_trial', yt.streams.first().default_filename)[:-4]+'mp4', os.path.join('videos_trial/', downloadURL.strip(BASE_URL)+'.mp4'))
            print("--------------->%s is loaded!" % name)
            duration=time.time()-t
            total_sec+=duration
            count+=1
            print('Running average time to process video: %f seconds'%(total_sec/count))
          except HTTPError as h:
            print(h)
            time.sleep(30)
            total_sec+=30
            continue   
      
def main():
    URL_set = get_downloadURLSet('videoIDs.txt')
    #in case running into recursion limit.
    #download_video(URL_set)
    T=time.time()
    cur_downloaded_size=len(list_current_IDs('videos_trial'))
    tot_vid=0
    #Set a for loop that sleeps 20 seconds and re-start the function after 
    #encoutering a non-HTTPError exception.
    #If it's an HTTPError, this simply means that we made too many requests 
    #to Youtube. Usually in this case we will just have to wait and restart.
    while True:
      try:
        download_video(URL_set)
      except:
        print('Error.Running again.')
        time.sleep(15)
        tot_vid_downloaded=len(list_current_IDs('videos_trial'))-cur_downloaded_size
        if tot_vid_downloaded>0:
          print('Total video processed %d'%tot_vid_downloaded)
          print('Total time to process video: %f minutes'%((time.time()-T)/60))   
          print('Average time to process video: %f seconds'%((time.time()-T)/tot_vid_downloaded))              
      else:
        tot_vid_downloaded=len(list_current_IDs('videos_trial'))-cur_downloaded_size
        print('Total time to process video: %f minutes'%((time.time()-T)/60))   
        print('Average time to process video: %f seconds'%((time.time()-T)/tot_vid_downloaded))              
        return

```
