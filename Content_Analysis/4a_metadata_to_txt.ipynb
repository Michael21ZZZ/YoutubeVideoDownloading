```python
UNLABEL_DATA_PATH = "./Metadata (Unlabeled Data) & Features Table/complete_metadata.csv"
LABEL_DATA_PATH = "./Metadata (Labeled data - Xiao) & Features Table/Data_Xiao.csv"
LABEL_TEXT_PATH = "./text/description_text/labeled/"
UNLABEL_TEXT_PATH = "./text/description_text/unlabeled/"
DESCRIPTION_TEXT_PATH = "./text/description_text/"
```

```python
import pandas as pd
```

```python
unlabeld_metadata = pd.read_csv(UNLABEL_DATA_PATH)
labeled_metadata = pd.read_csv(LABEL_DATA_PATH)
```

```python
unlabeld_metadata
```

```python
no_description = []
for row in unlabeld_metadata.itertuples():
    filename = UNLABEL_TEXT_PATH + row.id + ".txt"
    print(filename)
    if isinstance(row.description, str):
      with open(filename, 'w') as f:
        f.write(row.description)
    else:
      no_description.append(row.id)
```

```python
labeled_metadata
```

```python
no_description_label = []
for row in labeled_metadata.itertuples():
    filename = LABEL_TEXT_PATH + row.video_id + ".txt"
    print(filename)
    if isinstance(row.description, str):
      with open(filename, 'w') as f:
        f.write(row.description)
    else:
      no_description_label.append(row.video_id)
```

```python
with open(DESCRIPTION_TEXT_PATH + "notincluded.txt", 'w') as f:
  f.write("unlabeled: \n")
  f.write('\n'.join(no_description))
  f.write("\nlabeled: \n")
  f.write('\n'.join(no_description_label))
```
